```python
import imagend
import numpy as np
```

Generate some random sequence of images.

```python
image_sequence = np.random.rand(5, 20, 30, 40)
```

We can project just one stack...

```python
fig, axes = imagend.project_image_stack(image_sequence[0], add_labels=True);
fig.savefig("projection.png", bbox_inches='tight')
```

... or we can project the sequence.

```python
fig, axes = imagend.project_image_sequence(image_sequence, add_labels=True);
fig.savefig("sequence_of_projections.png", bbox_inches='tight')
```
