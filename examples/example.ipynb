```python
import imagend
import numpy as np
```

Generate some random sequence of images.

```python
image_sequence = np.random.rand(5, 20, 30, 40)
```

We can project just one stack...

```python
fig, axes = imagend.project_image_stack(image_sequence[0]);
fig.savefig("examples/projection.png")
```

... or we can project the sequence.

```python
fig, axes = imagend.project_image_sequence(image_sequence);
fig.savefig("examples/sequence_of_projections.png")
```
